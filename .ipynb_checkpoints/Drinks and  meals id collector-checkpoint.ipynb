```python
def drink(name):
    url = f"https://www.thecocktaildb.com/api/json/v1/1/search.php?s={str(name)}"
    r = requests.get(url)
    df = pd.DataFrame(r.json()["drinks"])
    return df.dropna(axis='columns', how='all')

def meal(name):
    url = f"https://www.themealdb.com/api/json/v1/1/search.php?s={str(name)}"
    r = requests.get(url)
    df = pd.DataFrame(r.json()["meals"])
    return df.dropna(axis='columns', how='all')
```

```python
drink("margarita")
```

```python
meal("chicken")
```

```python
def meal_ing(name):
    url = f"https://www.themealdb.com/api/json/v1/1/filter.php?i={str(name)}"
    r = requests.get(url)
    df = pd.DataFrame(r.json()["meals"])
    return df.dropna(axis='columns', how='all')
```

```python
meal_ing("chicken")
```

```python
def meal_id(name):
    url = f"https://www.themealdb.com/api/json/v1/1/lookup.php?i={name}"
    r = requests.get(url)
    df = pd.DataFrame(r.json()["meals"])
    return df.dropna(axis='columns', how='all')
```

```python
meal_id(52940)
```

```python
url ="https://www.themealdb.com/api/json/v1/1/filter.php?i=chicken_breast"
r = requests.get(url)
df = pd.DataFrame(r.json()["meals"])
```

```python
listid = df["idMeal"].values.tolist()
listid
```

```python
final = []
for id in listid:
    url =f"https://www.themealdb.com/api/json/v1/1/lookup.php?i={id}"
    r = requests.get(url)
    final.append(r.json()["meals"])

final2=[]
for x in final:
    final2.append(x[0])
pd.DataFrame(final2)
```
