## Drinks functions

### Choose drink options

``````python
#choosing between the options 
def drink_option_choose():
    print("""do you want to search by?
Name, Main ingredient, Category
    """)
    x = input().strip().lower()
    if x == "name" or x =="names":
        return drink_name_choose()
    elif x == "ingredient" or x == "ingredients" or x == "main" or x =="main ingredient" or x =="main ingredients":
        return drink_ingridient_choose()
    elif x == "category":
        return drink_category_choose()
    else:
        print("Sorry please choose a valide option")
        return drink_option_choose()
``````

### Drink name 

``````python
#Choose drink by name
def drink_name_choose():
    print("""Choose your drink name
    
    \033[92m
                   \ 
                   .\"""""""""-.
                   \`\-------'`/
                    \ \__ o . /
                     \/  \  o/
                      \__/. /
                       \_ _/
                         Y
                         |
                         |
                     _.-' '-._
                    `---------`
\033[0m
    """)
    x = input().strip().lower()
    return drink(x)

def drink(name):
    global ingridient_df
    url = f"https://www.thecocktaildb.com/api/json/v1/1/search.php?s={str(name)}"
    r = requests.get(url)
    df = pd.DataFrame(r.json()["drinks"])
    ingridient_df = df
    if df.empty == True:
        print("There are no matching names")
        drink_name_choose()
    else:
        #display(df.dropna(axis='columns', how='all'))
        display(df["strDrink"].to_frame().rename(columns = drinks_columns_dict))
    
    
    #changes
        if df.empty == True: 
            print("There are no matching drink names")
            return drink_name_choose()
        else:
            print("These are the items that match with your search. Please select a row to inspect.")
        
            x = input().strip().lower()
            try: 
                if int(x) <= len(ingridient_df):
                    return drink_id(x)
                elif int(x) > len(ingridient_df):
                    return drink(name)
            except:
                print("Please choose a number inside of the index")
                return drink(name)
    
``````

### Drink ingridient

``````python
#search by ingredient
def drink_ingridient_choose():
    print("""Choose your main ingridient
    """)
    x = input().strip().lower()
    
    
    #changes
    try:
        return drink_main(x)
    except: 
        print("There is no matching ingredient.")
        return drink_ingridient_choose()

def drink_main(main):
    global ingridient_df
    url = f"https://www.thecocktaildb.com/api/json/v1/1/filter.php?i={str(main)}"
    r = requests.get(url)
    df = pd.DataFrame(r.json()["drinks"])
    ingridient_df = df
    display(df.dropna(axis='columns', how='all')["strDrink"].to_frame().rename(columns = drinks_columns_dict))
    return drink_id_choose()    
``````

### Drink category


``````python
# Search for drink category
def drink_category_choose():
    global category_df
    print("""this are the categories please choose the row you want to inspect
    """)
    url_list_origin = "https://www.thecocktaildb.com/api/json/v1/1/list.php?c=list"
    ra = requests.get(url_list_origin)
    df = pd.DataFrame(ra.json()["drinks"])
    category_df = df
    display(df.rename(columns = drinks_columns_dict))
    x = input().strip().lower()
    try: 
        if int(x) <= len(category_df):
            return drink_category_row(x)
        elif int(x) > len(category_df):
            return drink_category_choose()
    except:
        print("Please choose a number inside of the index")
        return drink_category_choose()

def drink_category_row(x):
    global category_df
    global ingridient_df
    category = category_df.loc[int(x), "strCategory"]
    url = f"https://www.thecocktaildb.com/api/json/v1/1/filter.php?c={str(category)}"
    r = requests.get(url)
    dfcat = pd.DataFrame(r.json()["drinks"])
    ingridient_df = dfcat
    display(dfcat.dropna(axis='columns', how='all').rename(columns = drinks_columns_dict).drop(columns = ["strDrinkThumb", "idDrink"]))#.drop(columns=['strInstructions','strInstructionsDE','strCreativeCommonsConfirmed'], axis=1).transpose()
    return drink_id_choose()
``````

### Drink id

``````python
#All the other functions call this one in the end for serching the drink with its id
def drink_id_choose():
    print("""Choose the row you want to examine
    """)
    x = input().strip().lower()
    #changes
    try: 
        if int(x) <= len(ingridient_df):
            return drink_id(x)
        elif int(x) > len(ingridient_df):
            print("please choose a number inside of the index")
            return drink_id_choose()
    except:
        print("please choose a number inside of the index")
        return drink_id_choose()

def drink_id(x):
    global drinks_columns_dict
    global drink_columns_rearranged 
    global ingridient_df
    id = ingridient_df.loc[int(x), "idDrink"]
    url = f"https://www.thecocktaildb.com/api/json/v1/1/lookup.php?i={id}"
    r = requests.get(url)
    dfid = pd.DataFrame(r.json()["drinks"])
    dfid = dfid[drink_columns_rearranged].rename(columns = drinks_columns_dict)
    display(dfid["Instructions"].str.replace('\n', '').str.replace('\r', '').to_frame())
    display(dfid.dropna(axis='columns', how='all').drop(columns=["strDrinkThumb", "dateModified",'Instructions','strInstructionsDE','strCreativeCommonsConfirmed'], axis=1).transpose())
    return restart()
``````

### Strip white spaces

``````python
# function to clean empty columns
def drop_empty_columns(df):
    ls = []
    df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
    for i in list(df.columns):
        if df[i].any() == '':
            ls.append(i)
    return df.drop(columns = ls)
``````

### Meal id

``````python
#All the other functions call this one in the end for serching the meal with its id
def meal_id_choose():
    global ingridient_df
    print("""Choose the row you want to examine
    """)
    x = input().strip().lower()
    #changes
    try: 
        if int(x) <= len(ingridient_df):
            return meal_id(x)
        elif int(x) > len(ingridient_df):
            print("please choose a number inside of the index")
            return meal_id_choose()
    except:
        print("please choose a number inside of the index")
        return meal_id_choose()
    
def meal_id(x):
    global meal_columns_dict
    global meal_columns_rearranged
    global ingridient_df
    id = ingridient_df.loc[int(x), "idMeal"]
    url = f"https://www.themealdb.com/api/json/v1/1/lookup.php?i={id}"
    r = requests.get(url)
    dfid = pd.DataFrame(r.json()["meals"])
    dfid = dfid[meal_columns_rearranged].rename(columns = meal_columns_dict)
    display(dfid["Instructions"].str.replace('\n', '').str.replace('\r', '').to_frame())
    display(drop_empty_columns(dfid).drop(columns=['Instructions','strMealThumb','strYoutube','strSource'], axis=1).dropna(axis='columns', how='all').transpose())
    return restart()
``````

## Restart or break

``````python
#Restarts the functions and resets the global varibles
def restart():
    global ingridient_df
    global category_df
    global area_df
    print("do you want to search another drink or meal? (Yes/No)") #changes
    x = input().strip().lower()
    if x == "yes":
        ingridient_df = ""
        category_df = ""
        area_df = ""
        return start()
    elif x == "no":
        print("Have a nice evening")
    else:
        return restart()
``````
