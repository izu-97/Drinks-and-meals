## Function to get user selection of the glass of the drink and provide list of options


```python
import pandas as pd
import requests
import json
import matplotlib
#pd.options.display.max_rows = 999
#pd.options.display.max_columns = 999

%matplotlib inline
```

```python
# Search for drink category
def drink_glass(glass):
    url = f"https://www.thecocktaildb.com/api/json/v1/1/filter.php?g={str(glass)}"
    r = requests.get(url)
    df = pd.DataFrame(r.json()["drinks"])
    return df.dropna(axis='columns', how='all')
```

```python
# for a given name, requires to replace " " by "_"
drink_glass("Punch_bowl")
```

```python
# needs to be exact
# list of different categories for this db
url_list_origin = "https://www.thecocktaildb.com/api/json/v1/1/list.php?g=list"
ra = requests.get(url_list_origin)
df = pd.DataFrame(ra.json()["drinks"])
df
```
