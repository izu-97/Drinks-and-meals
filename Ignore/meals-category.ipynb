## Function to get user selection of the category of the food and provide list of options


```python
import pandas as pd
import requests
import json
import matplotlib
#pd.options.display.max_rows = 999
#pd.options.display.max_columns = 999

%matplotlib inline
```

```python
# this function returns the list of the meals given origin
def meal_category(category):
    url = f"https://www.themealdb.com/api/json/v1/1/filter.php?c={str(category)}"
    r = requests.get(url)
    df = pd.DataFrame(r.json()["meals"])
    return df.dropna(axis='columns', how='all')
```

```python
# testing the function works
meal_category("Seafood")
```

```python
# Check the different values of the area
url_list_origin = "https://www.themealdb.com/api/json/v1/1/list.php?c=list"
ra = requests.get(url_list_origin)
df = pd.DataFrame(ra.json()["meals"])
df
```
