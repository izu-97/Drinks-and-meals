## Drinks functions

### Choose drink options

```python
def drink_option_choose():
    print("""do you want to search by?
Name, Main ingridient, Category
    """)
    x = input().strip().lower()
    if x == "name":
        return drink_name_choose()
    elif x == "ingridient":
        return drink_ingridient_choose()
    elif x == "category":
        return drink_category_choose()
    else:
        print("Sorry please choose a valide option")
        return drink_option_choose()
```

### Drink name 

```python
def drink_name_choose():
    print("""Choose your drink name
    """)
    x = input().strip().lower()
    return drink(x)

def drink(name):
    global ingridient_df
    url = f"https://www.thecocktaildb.com/api/json/v1/1/search.php?s={str(name)}"
    r = requests.get(url)
    df = pd.DataFrame(r.json()["drinks"])
    ingridient_df = df
    if df.empty == True:
        print("There are no matching names")
        drink_name_choose()
    else:
        #display(df.dropna(axis='columns', how='all'))
        display(df["strDrink"].to_frame().rename(columns = drinks_columns_dict))
    
    
    #changes
        if df.empty == True: 
            print("There are no matching drink names")
            return drink_name_choose()
        else:
            print("These are the items that match with your search. Please select a row to inspect.")
        
            x = input().strip().lower()
            try: 
                if int(x) <= len(ingridient_df):
                    return drink_id(x)
                elif int(x) > len(ingridient_df):
                    return drink(name)
            except:
                print("Please choose a number inside of the index")
                return drink(name)
    
```

### Drink ingridient

```python
def drink_ingridient_choose():
    print("""Choose your main ingridient
    """)
    x = input().strip().lower()
    
    
    #changes
    try:
        return drink_main(x)
    except: 
        print("There is no matching ingredient.")
        return drink_ingridient_choose()

def drink_main(main):
    global ingridient_df
    url = f"https://www.thecocktaildb.com/api/json/v1/1/filter.php?i={str(main)}"
    r = requests.get(url)
    df = pd.DataFrame(r.json()["drinks"])
    ingridient_df = df
    display(df.dropna(axis='columns', how='all')["strDrink"].to_frame().rename(columns = drinks_columns_dict))
    return drink_id_choose()    
```

### Drink category


```python
# Search for drink category
def drink_category_choose():
    global category_df
    print("""this are the categories please choose the row you want to inspect
    """)
    url_list_origin = "https://www.thecocktaildb.com/api/json/v1/1/list.php?c=list"
    ra = requests.get(url_list_origin)
    df = pd.DataFrame(ra.json()["drinks"])
    category_df = df
    display(df.rename(columns = drinks_columns_dict))
    x = input().strip().lower()
    try: 
        if int(x) <= len(category_df):
            return drink_category_row(x)
        elif int(x) > len(category_df):
            return drink_category_choose()
    except:
        print("Please choose a number inside of the index")
        return drink_category_choose()

def drink_category_row(x):
    global category_df
    global ingridient_df
    category = category_df.loc[int(x), "strCategory"]
    url = f"https://www.thecocktaildb.com/api/json/v1/1/filter.php?c={str(category)}"
    r = requests.get(url)
    dfcat = pd.DataFrame(r.json()["drinks"])
    ingridient_df = dfcat
    display(dfcat.dropna(axis='columns', how='all').rename(columns = drinks_columns_dict).drop(columns = ["strDrinkThumb", "idDrink"]))#.drop(columns=['strInstructions','strInstructionsDE','strCreativeCommonsConfirmed'], axis=1).transpose()
    return drink_id_choose()
```

### Strip white spaces

```python
# function to clean empty columns
def drop_empty_columns(df):
    ls = []
    df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
    for i in list(df.columns):
        if df[i].any() == '':
            ls.append(i)
    return df.drop(columns = ls)
```

## Restart or break

```python
def restart():
    global ingridient_df
    global category_df
    global area_df
    print("do you want to restart the game? (Yes/No)") #changes
    x = input().strip().lower()
    if x == "yes":
        ingridient_df = ""
        category_df = ""
        area_df = ""
        return start()
    elif x == "no":
        print("Have a nice evening")
    else:
        return restart()
```
