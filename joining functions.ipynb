```python
import pandas as pd
import requests
import json
import matplotlib
pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
pd.set_option("display.max_colwidth", -1)

%matplotlib inline
```

```python
ingridient_df = ""
```

```python
def drink_id(x):
    id = ingridient_df.loc[int(x), "idDrink"]
    url = f"https://www.thecocktaildb.com/api/json/v1/1/lookup.php?i={id}"
    r = requests.get(url)
    dfid = pd.DataFrame(r.json()["drinks"])
    display(dfid["strInstructions"].to_frame())
    return dfid.dropna(axis='columns', how='all').drop(columns=['strInstructions','strInstructionsDE','strCreativeCommonsConfirmed'], axis=1).transpose()
```

```python
def drink_id_choose():
    global ingridient
    print("""Choose the row you want to examine
    """)
    x = input().strip().lower()
    return drink_id(x)
```

```python
def drink_name_choose():
    print("""Choose your name
    """)
    x = input().strip().lower()
    return drink(x)
```

```python
def drink_ingridient_choose():
    global ingridient
    print("""Choose your main ingridient
    """)
    x = input().strip().lower()
    return drink_main(x)
```

```python
def drink_main(main):
    global ingridient_df
    url = f"https://www.thecocktaildb.com/api/json/v1/1/filter.php?i={str(main)}"
    r = requests.get(url)
    df = pd.DataFrame(r.json()["drinks"])
    display(df.dropna(axis='columns', how='all')["strDrink"].to_frame())
    ingridient_df = df
    return drink_id_choose()
    
```

```python
def drink_option_choose():
    print("""do you want to search by?
Name, Ingridient, Type, Glass
    """)
    x = input("Drink").strip().lower()
    if x == "name":
        return drink_name_choose()
    if x == "ingridient":
        return drink_ingridient_choose()
```

```python
def meal_choose():
    print("what do you want to eat?")
    x = input("meal").strip().lower()
    return meal(x)
    
```

```python
def meal(name):
    url = f"https://www.themealdb.com/api/json/v1/1/search.php?s={str(name)}"
    r = requests.get(url)
    df = pd.DataFrame(r.json()["meals"])
    return df.dropna(axis='columns', how='all')
```

```python
def drink(name):
    url = f"https://www.thecocktaildb.com/api/json/v1/1/search.php?s={str(name)}"
    r = requests.get(url)
    df = pd.DataFrame(r.json()["drinks"])
    return df.dropna(axis='columns', how='all')
```

# This functions starts everything

```python
#calling the other functions

def start():
    print("welome to your night what do you want to do drink or meal?")
    x = input().strip().lower()
    if x == "drinks":
        return drink_option_choose()
    elif x == "meal":
        return meal_choose()
    return meal()
```
