```python
import pandas as pd
import requests
import json
```

```python
# function get meal based on id for testing
def get_meal_id(id):
    url = f"https://www.themealdb.com/api/json/v1/1/lookup.php?i={id}"
    r = requests.get(url)
    dfid = pd.DataFrame(r.json()["meals"])
    return dfid.dropna(axis='columns', how='all')
```

```python
# get one meal as example
df = get_meal_id(52795)
```

```python
# function to get a drink dataset
def drink_id(id):
    url = f"https://www.thecocktaildb.com/api/json/v1/1/lookup.php?i={id}"
    r = requests.get(url)
    dfid = pd.DataFrame(r.json()["drinks"])
    return dfid.dropna(axis='columns', how='all').drop(columns=['strInstructions','strInstructionsDE','strCreativeCommonsConfirmed'], axis=1)
```

```python
# get one drink as example
df = drink_id(11007)
```
